# Phase 0 — Reconnaissance & Cadrage métier

Cette phase zéro a pour but de poser le cadre métier et de générer des impressions et pistes d'exploration
 
Avant tout travail concret sur le dataset lui-même



## Contexte du dataset : 

Ensemble de données qui représentent les ventes
d'un commerce de détail en ligne

L'ensemble des transactions est évalué en livres sterling au niveaux des prix

donc le travail d'analyse prendra en compte ce paramètre.


## Origine : 

Les données proviennent d'un contexte de simulation du métier de data analyst

De l'entreprise TATA sur la plateforme Forage.

Les données en question ont été mises à disposition sur la plateforme Kaggle.


## Problématique :

Je joue le rôle de consultant data auprès de la boutique en ligne qui a sollicité mes services

pour les aider à identifier les principaux moteurs de croissance du chiffre d'affaires ainsi qu'à élaborer une stratégie pour l'année suivante.


## Dictionnaire de chaque colonne
| Colonne | Utilité analytique |
|---|---|
| InvoiceNo | Identifiant de transaction — préfixe C = annulation |
| StockCode | Clé produit unique |
| Description | Nom lisible du produit |
| Quantity | Volume — négatif sur annulations |
| InvoiceDate | Tendances temporelles |
| UnitPrice | Prix en £ — calcul du CA |
| CustomerID | Identifiant client — valeurs manquantes |
| Country | Dimension géographique |

## Questions business (3 max) : 

1- Analyse des revenus : Quels produits, périodes et zones géographiques génèrent le plus de chiffre d'affaires ?

2- Comportement client: Quels clients sont les plus actifs et comment se segmentent-ils par fréquence et valeur d'achat ?

3- Tendances & anomalies :Quelles tendances temporelles observe-t-on et quelles anomalies dans les données impactent la fiabilité de l'analyse ?


## KPIs à suivre :

* CA total, CA par pays, CA par produit
* Nombre de transactions par période
* Fréquence d'achat par client
* Valeur moyenne par commande
* Taux de retours / annulations (valeurs négatives dans Quantity)


## Limites du dataset :

1- Il ny'a pas de catégories de produits ce qui limite l'analyse de ce côté

2- tout ce qui est indiqué sur les clients c'est la colonne customerID, et sur cette même colonne , il y a des données manquantes
On n'a pas les âges, pas les genres, la segmentation est limitée.

3- on ne peut pas pousser l'analyse plus loin sur l'efficacité marketing car il n'y a aucune colonne correspondantes
au métriques associées comme le taux de conversion ou les canaux de vente.


## Hypothèses initiales 

-Les clients du Royaume-Uni peuvent être ceux qui effectuent le plus de transactions.

-Les périodes de fin d'année doivent être celles où le chiffre d'affaires est augmenté
Par rapport aux périodes plus creuses comme la première moitié de l'année.

-le taux d'abandon de panier semble très faible par rapport aux achats finalisés 

## Bilan

Le but d'une boutique en ligne c'est de faire venir des gens, les convaincre d'acheter , puis les fidéliser

Cette phase de reconnaissance pose le cadre métier de l'étude : 
Comprendre les données disponibles, définir les questions auxquelles on peut répondre et fixer les limites de l'analyse.

Les trois questions business retenues — revenus, comportement client et tendances de ventes — sont des pistes concrètes pour aider la boutique à prendre des décisions et booster son chiffre d'affaires.

Une analyse de l'efficacité marketing n'a pas pu être menée : 
Le dataset ne contient aucune donnée sur les canaux de vente ni les taux de conversion.
C'est une limite du dataset, pas de l'analyse.
