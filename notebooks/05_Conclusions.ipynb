# Phase 6 — Conclusions et recommandations

Cette phase vient clôturer le travail d'investigations pour répondre aux problématiques de la boutique

Elle présentera directement ce que le client veut savoir, et doit mettre en place

---
# conntexte de missions
---

Le consultant data a été mandaté pour :
 - identifier les facteurs qui contribuent à la croissance du chiffre d'affaires de la boutique, 
 - et implicitement , les facteurs qui le freinent.
 
 L'objectif est d'en tirer une stratégie concrète pour l'année suivante : 
 > *amplifier ce qui performe, et décider quoi faire de ce qui pose problème.*

---
# ce qui a été découvert:
---
    
### Quels produits et périodes génèrent le plus de revenus ?

**REGENCY CAKESTAND 3 TIER** est le produit le plus lucratif du catalogue. Il le doit à son volume de commandes. Mais c'est aussi celui qui affiche le taux d'annulation le plus élevé parmi les 20 premiers produits (8.22%). Un signal à surveiller.

**PICNIC BASKET WICKER 60 PIECES** est un cas à part : dépense moyenne par commande très élevée, zéro annulation, mais volume faible. Un produit qui pourrait rapporter beaucoup plus s'il était davantage mis en avant.

**Les mois de septembre à novembre** sont les plus lucratifs, avec un CA mensuel dépassant le million de livres sterling. **Avril** est le point bas de l'année.


### Qui sont vos meilleurs clients et d'où viennent-ils ?

**Le Royaume-Uni représente 84.7% du chiffre d'affaires global**. La boutique est avant tout une boutique britannique, les marchés internationaux restent marginaux en valeur comme en volume.
**Les clients non identifiés** constituent le segment le plus actif : ils génèrent le plus de CA et commandent le plus fréquemment. Leur taux d'annulation est quasi nul **(0.12%)**, ce qui en fait un profil fiable malgré l'absence d'information. **À Hong Kong, ils représentent 100% de la clientèle.**

Dans l'ensemble, les clients qui commandent le plus souvent ne sont pas ceux qui rapportent le plus en valeur 
**La fréquence et la valeur, en ce qui concerne les clients, ne sont pas en parfaite corrélation**


### Quelles tendances et anomalies faut-il surveiller ?

La saisonnalité est marquée et prévisible : **les revenus progressent nettement à partir de septembre**, portés par les achats de fin d'année. Le début d'année est structurellement creux.
**Les USA** affichent un taux d'annulation de 38.62% pour seulement 7 commandes et un CA négligeable, **c'est un marché peu développé et peu fiable.**
Le **jeudi** concentre le plus grand nombre de commandes, et le **dimanche**, le volume est deux fois plus faible.

# Interpretation des constats
---

### 1 - dimensions temporelles:

**Constat** : Les mois `septembre–novembre` concentrent les pics de CA. Le `jeudi` est le jour le plus actif, le dimanche le plus creux : volume deux fois plus faible.

**Comportement détecté:**
Les clients dépensent plus au moment où les périodes de fin d'année commencent à s'annoncer. En semaine, le volume de commandes est  plus élevé avec le jeudi en tête. Le week-end, l'activité chute, avec un volume deux fois plus faible le dimanche.

**Risque ou opportunité ?**

Opportunité : la saisonnalité est factuelle, mais comme le dataset se base sur  l'année 2011, on ne peut pas s'avancer en le nommant de comportement prévisible
Risque : le début d'année  creux, sans action corrective le CA s'effondre mécaniquement au début de janvier

**Action recommandée:**

- Anticiper les stocks et la logistique dès août pour absorber le pic de septembre–novembre sans rupture.

 - Pour comprendre et potentiellement atténuer le creux de début d'année, une investigation côté clients serait nécessaire : qui commande en    janvier–avril, et pourquoi leur comportement diffère-t-il de celui de la fin d'année ? Les données actuelles ne permettent pas de répondre à cette question.

 - Concentrer les communications marketing en milieu de semaine : le jeudi en priorité ,  pour coller aux habitudes d'achat des clients.

---
### 2 - dimensions clients:

**constat:** 
Les `clients inconnus` tirent le CA vers le haut, ces mêmes clients n'annulent pas beaucoup non plus : comme 1 seul client Unknown avec 276 transactions à Hong Kong. C'est un seul profil non identifié très actif. Profil compatible avec un revendeur.

**Comportement client:**
La valeur d’achat la plus grande est `anonyme`. Ces clients ne créent peut-être pas de compte avant de commander. Est-ce qu’ils sont juste de passage pour des achats ponctuels récurrents ? Parmi les clients identifiés, ceux  qui commandent le plus souvent ne sont pas ceux qui rapportent le plus en valeur.

**Opportunités ou risques ?**
- Opportunités: les clients inconnus sont rentables et fiables
- Risque:  par contre, ces mêmes clients sont non traçables, donc non fidélisables

**Les actions recommandées deviennent donc deux actions distinctes :**

  - Savoir les identifier sans les perdre avec 3 options, chacune des options comporte un risque business:
    1- une création de compte obligatoire, on gagne une identification, mais certains peuvent refuser la contrainte
    2- un compte fortement encouragé, on gagne une part d'identification, mais aussi une possibilité de segment intraçable
    3- mettre en place un avantage à avoir un compte client ( une remise, des recommandations de produits similaires après achat), mais ça peut impliquer une perte acceptable de revenus.

  - pour  les clients identifiés à haute valeur mais faible fréquence :
Comprendre pourquoi certains clients dépensent beaucoup en une seule commande puis disparaissent. Sont-ils des acheteurs occasionnels, des revendeurs ponctuels ? Mettre en place un questionnaire de satisfaction qui pourrait indiquer pourquoi ils ne sont pas fréquents.


---
### 3 - dimensions produit:

**Constat:**
Les produits qui génèrent le plus de CA le doivent à leur volume de commandes, ils apparaissent dans un grand nombre de transactions. 
À l'opposé, certains produits comme le `PICNIC BASKET WICKER 60 PIECES` affichent une valeur par transaction très élevée, mais un volume faible.

dans ce dataset, Le `PICNIC BASKET WICKER 60 PIECES` ne repose que sur 2 transactions d'un seul client :
 - une commande test d'une unité 
 - suivie d'une commande en volume de 60 unités.
Son profil est compatible avec un acheteur professionnel ou un revendeur. Zéro annulation, prix unitaire élevé à `649 £` : si d'autres acheteurs de ce profil existent, le potentiel de CA est réel

Deux comportements d'achat distincts coexistent dans le catalogue.

Le `REGENCY CAKESTAND 3 TIER`, premier en CA, présente également le taux d'annulation le plus élevé parmi les 20 produits les plus lucratifs (8.22%).

**Comportement détecté:**
`Les clients achètent régulièrement des produits à fort volume`, c'est la fréquence d'achat qui construit le CA, pas la valeur unitaire. 
Les produits à forte valeur unitaire restent peu commandés. La question à examiner : est-ce leur prix qui freine la demande ? le prix élevé est une hypothèse plausible pour expliquer le faible volume, mais non vérifiable avec ce dataset.

**Risque ou opportunité:**
Risque : Le CA repose sur un petit nombre de produits à fort volume. Si la popularité de l'un d'eux diminue, l'impact sur le CA global sera immédiat..
Opportunité : Le PICNIC BASKET représente une opportunité non exploitée : fort potentiel de valeur, zéro annulation. Ce que les données ne permettent pas de dire : pourquoi se vend-il peu ?

**Action recommandée**
- Investiguer les causes des annulations sur le REGENCY CAKESTAND avant de tirer des conclusions, sans cette information, aucune action corrective n'est pertinente.
- Tester la visibilité du PICNIC BASKET : est-il suffisamment mis en avant dans le catalogue ? Une meilleure exposition entraînerait le type de clientèle associé ( revendeur ou acheteur professionnel) à se manifester davantage, ce qui mènerait à plus de commandes.

---
### 4 - dimensions géographiques:

**Constat:**
le marché international est laissé à la marginalité face à la forte concentration du  marché local britannique. 

**Comportement détecté** : Le `Royaume-Uni` est le pays qui tire le CA vers le haut, et le plus actif en fréquence de commandes

**Risque ou opportunité:**
Opportunité: la boutique peut très bien miser là-dessus, sur son marché local qui rapporte déjà très bien 
Risque : mettre tous ses œufs dans le même panier n'est pas toujours la méthode la plus fiable, si un jour il y a une baisse d'activité, sans marché secondaire où la boutique pourrait rebondir . Le risque de concentration est donc à éviter

**actions recommandés:**
Est-ce qu’à la base, la boutique avait prévu de se concentrer uniquement sur le marché britannique, ou bien l’expansion à l’international a échoué ?
maintenir UK en top 1 n'est pas viable si le CA est aussi écrasant, développer des marchés secondaires pour réduire la dépendance au Royaume-Uni

# Recommendations par top priorité
---

## 1. Sécuriser les produits qui portent le CA

Investiguer les causes d'annulation sur le `REGENCY CAKESTAND 3 TIER` , premier produit en CA, mais aussi premier en taux d'annulation parmi les 20 produits les plus lucratifs. 
Sans cette investigation, le risque reste non géré. 
En parallèle, tester la visibilité du `PICNIC BASKET WICKER 60 PIECES` dans le catalogue présentant un fort potentiel de valeur, zéro annulation, volume trop faible pour son profil. investiguer aussi sur son type de clientèle


## 2. Réduire la dépendance au marché britannique

`84.7% du CA concentré sur un seul marché durant l'année 2011`,  est un risque de concentration élevé. 
Avant toute action internationale, clarifier l'intention stratégique : la boutique a-t-elle vocation à rester locale ou à se développer à l'international ? 

   
## 3. Identifier les clients Unknown

Prendre une décision d'identification en fonction de ce qu'on peut minimiser comme risques acceptables.
Ces clients génèrent le plus de CA, annulent peu et sont présents sur des marchés internationaux où ils représentent parfois la totalité de la clientèle (exemple de Hong Kong). 
Les identifier est un prérequis au ciblage marketing.


## 4. Investiguer les clients à haute valeur, faible fréquence

Ces clients dépensent beaucoup en une seule commande puis disparaissent. 
Comprendre pourquoi ils ne reviennent pas pourrait débloquer une source de CA supplémentaire sans acquisition de nouveaux clients.


## 5. Comprendre les creux temporels

Investiguer le profil des clients en début d'année. Cette investigation sera plus productive une fois les clients Unknown identifiés 
les données actuelles ne permettent pas de répondre à cette question de manière fiable.

# Limites & points de vigilance
---

### Limites des données

| # | Limite | Conséquence |
|---|--------|-------------|
| 1 | Décembres tronqués (2010 : 23 jours, 2011 : 8 jours) | La période la plus lucrative de l'année reste non mesurable, le CA réel de décembre est inconnu |
| 2 | Une seule année complète analysée (2011) | Impossible de qualifier les tendances de saisonnalité comme récurrentes |
| 3 | Absence de catégories client | Impossible de qualifier le profil des acheteurs : particuliers, professionnels, revendeurs |
| 4 | Absence de données d'acquisition client | On ne sait pas d'où viennent les clients ni comment ils ont découvert la boutique |
| 5 | Double incertitude Unknown / Unspecified | Deux zones d'ombre sur le même flux financier non exploitable stratégiquement |

### Limites de l'analyse

| # | Limite | Conséquence |
|---|--------|-------------|
| 1 | Valeur moyenne par commande calculée mais non croisée | La moyenne de 517.88£ masque probablement une distribution asymétrique : quelques commandes à très haute valeur tirent la moyenne vers le haut. Sans médiane ni distribution complète calculées, ce chiffre doit être interprété avec prudence. La métrique reste un signal utile, mais insuffisant seul. |
| 2 | Taux d'annulation calculé sur les lignes de transactions pour les dimensions clients et pays: un calcul sur les factures distinctes serait plus rigoureux | Surestimation possible sur les clients et pays |
| 3 | `nb_commandes` utilisé comme approximation du volume produit à la place de `quantity` | Mesure le reach du produit : dans combien de commandes il apparaît , mais pas le volume réel d'unités commandées avec intention d'achat |
| 4 | Aucun top produit n'atteint 2% du CA global |Même les 20 produits les plus lucratifs restent chacun en dessous de 2% du CA global. Aucun produit ne joue le rôle de locomotive |
| 5 | Ratio 20 transactions par commande stable toute l'année | Probablement artificiel dans le dataset : peu exploitable pour des recommandations réelles |

# Conclusions
---

Cette analyse a mis en évidence quatre facteurs qui portent le CA de la boutique : 

`les périodes de fin d'année`

`les clients inconnus`

`le Royaume-Uni`

`les produits à fort volume de commandes, indépendamment de leur valeur unitaire.`

Les produits restent le levier le plus direct sur les revenus. Deux signaux méritent une attention immédiate : 

 - le taux d'annulation du **REGENCY CAKESTAND 3 TIER**, premier produit en CA, qui doit être investigué et optimisé ; 
 - Et le **PICNIC BASKET WICKER 60 PIECES**, fort potentiel de CA, zéro annulation, mais volume trop faible pour son profil.

Pour la suite, la question stratégique centrale est géographique : 
> *la boutique doit décider si elle consolide son marché britannique ou engage une diversification internationale. Cette décision conditionne l'ensemble des actions à venir.*
